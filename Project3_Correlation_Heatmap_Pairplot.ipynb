## Dataset Description

The Superstore dataset contains sales records including information about
sales, profit, quantity, and discount. These variables help analyze business
performance and relationships between different numerical features.

```python
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

## Selecting Numerical Features

For correlation analysis, only numerical variables are selected.  
The features used in this project are:

- Sales
- Profit
- Quantity
- Discount

```python
data = df[['Sales','Profit','Quantity','Discount']]
data.head()
```

## Pearson Correlation Matrix

Pearson correlation measures the strength and direction of the linear
relationship between numerical variables. Values range from -1 to 1.

- **1** → Strong positive correlation  
- **0** → No correlation  
- **-1** → Strong negative correlation

```python
corr = data.corr()
corr
```

## Correlation Heatmap

A heatmap visually represents the correlation matrix using colors.  
Stronger correlations appear with darker colors.  

To improve readability, the upper triangle of the matrix is masked so
that only unique correlations are displayed.

```python
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(6,4))

sns.heatmap(corr,
            mask=mask,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            fmt=".2f")

plt.title("Correlation Heatmap")
plt.savefig("heatmap.png")

plt.show()
```

## Pairwise Relationships (Pairplot)

Pairplots show scatter plots for each pair of variables along with the
distribution of individual variables. This helps identify patterns,
trends, and potential relationships between variables.

```python
sns.pairplot(data, diag_kind="kde")

plt.savefig("pairplot.png")

plt.show()
```

## Conclusion

The correlation heatmap shows the relationships between numerical variables
in the dataset. Sales and Profit show a positive correlation, indicating that
higher sales generally lead to increased profit. Discount shows a negative
correlation with Profit, suggesting that higher discounts may reduce
profitability. Pairplots further illustrate the relationships and
distribution patterns between variables, helping to better understand how
these features interact.
